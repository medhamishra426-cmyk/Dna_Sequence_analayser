```python
import pandas as pd
#create a simple dataframe
data = {
    "Name": ["Adenine","Thymine","Guanine","Cytosine"],
    "Symbol": ["A","T","G","C"],
    "Type" : ["Purine","Pyrimidine","Purine","Pyrimidine"]

}

df = pd.DataFrame(data)
print(df)
print(df.describe())
print(df["Type"].value_counts())
```

```python
#DNA sequence Analyser
#step 1 - take a DNA sequence as input

dna = input("Enter a DNA sequence:").upper()
print("You entered:", dna)
```

```python
#step 2 - count each nucleotide
count_A = dna.count("A")
count_T = dna.count("T")
count_G = dna.count("G")
count_C = dna.count("C")

print("Adenine (A):", count_A)
print("Thymine (T):", count_T)
print("Guanine (G):", count_G)
print("Cytosine (C):", count_C)
```

```python
#step 3 - calculate gc content

total = len(dna)
gc_content = (count_G + count_C) / total * 100

print("Total length:", total)
print("Gc content:", round(gc_content, 2), "%")
```

```python
#step 4 - validate the sequence

valid_bases = set ("ATGC")
input_bases = set(dna)

if input_bases.issubset(valid_bases):
  print("valid DNA sequence!")
else:
    invalid = input_bases - valid_bases
    print("invalid sequence! Found these invalid characters:", invalid)
```

```python
#step 5 - final summary output

print("=" * 30)
print("  DNA SEQUENCE ANALYSER")
print("=" * 30)
print("sequence:", dna)
print("length:", total)
print("-"* 30)
print("Nucleotide Counts:")
print(" A (adenine) :", count_A)
print(" T (thymine) :", count_T)
print(" G (guanine) :", count_G)
print(" C (cytosine) :", count_C)
print("-"* 30)
print("GC Content  :", round(gc_content,2), "%")
print("=" * 30)
```
